## 4. Features

**Create data dictionary**

1. **age**  
Patient's age in years. Represents the chronological age of the individual at the time of examination, which is a significant risk factor for cardiovascular disease.

2. **sex**  
Biological sex of the patient. Encoded as binary variable where 0 represents female and 1 represents male. Gender differences play an important role in heart disease risk assessment.

3. **cp (chest pain type)**  
Type of chest pain experienced by the patient, categorized into four distinct types:
- 0: Asymptomatic (no chest pain symptoms)
- 1: Typical angina (classic chest pain related to reduced blood flow to heart)
- 2: Atypical angina (chest pain not fully matching typical angina characteristics)
- 3: Non-anginal pain (chest pain not related to heart condition)

4. **trestbps (resting blood pressure)**  
Resting systolic blood pressure measured in mmHg when the patient was admitted to the hospital. Higher values indicate increased cardiovascular risk.

5. **chol (serum cholesterol)**  
Total serum cholesterol level measured in mg/dl. Elevated cholesterol levels are associated with increased risk of coronary artery disease.

6. **fbs (fasting blood sugar)**  
Binary indicator of whether fasting blood glucose level exceeds 120 mg/dl. Value of 1 indicates fasting blood sugar > 120 mg/dl (suggesting diabetes risk), while 0 indicates â‰¤ 120 mg/dl.

7. **restecg (resting electrocardiographic results)**  
Results from resting electrocardiogram showing electrical activity of the heart:
- 0: Normal ECG reading
- 1: ST-T wave abnormality (indicating possible heart problems)
- 2: Left ventricular hypertrophy (enlarged heart muscle)

8. **thalach (maximum heart rate achieved)**  
Maximum heart rate achieved during exercise stress test, measured in beats per minute. Lower maximum heart rates may indicate cardiovascular limitations.

9. **exang (exercise induced angina)**  
Binary variable indicating whether exercise induces chest pain (angina). Value of 1 means yes (exercise-induced angina present), 0 means no angina during exercise.

10. **oldpeak**  
ST depression induced by exercise relative to rest, measured in millimeters. This represents the difference in ST segment depression between exercise and rest states on the ECG. Higher values may indicate coronary artery disease.

11. **slope**  
The slope of the peak exercise ST segment on the electrocardiogram:
- 0: Downsloping (often associated with heart disease)
- 1: Flat (may indicate heart problems)
- 2: Upsloping (typically normal response)

12. **ca (number of major vessels)**  
Number of major coronary vessels (0-3) colored by fluoroscopy, indicating the extent of coronary artery blockage. Higher numbers suggest more severe coronary artery disease.

13. **thal (thalassemia)**  
Results from thallium stress test showing blood flow to the heart:
- 0: Normal blood flow
- 1: Fixed defect (permanent blood flow reduction)
- 2: Reversible defect (temporary blood flow reduction during stress)

14. **Target**  
Whether (1) or not (0) the patient has a heart disease

## 5. Modeling

```python
df.head()
```

```python
# Split data into X and y
X = df.drop("target", axis = 1)

y = df["target"]
```

```python
# Split data into train and test sets
np.random.seed(42)

# Split into train and test sets
X_train, X_test, y_train, y_test = train_test_split(X , y, test_size = 0.2 )
```

Trying 3 different ML models:
1. Logistic Regression
2. K-Nearest Neighbours Classifier
3. Random Forest Classifier

```python
# Put models into a dictionary
models = { "Logistic Regression" : LogisticRegression(),
          "KNN" : KNeighborsClassifier(),
          "Random Forest" : RandomForestClassifier()
}

# Create a function to fit and scores models
def fit_and_score(models, X_train, X_test, y_train, y_test):
    """
    Fits and evaluates given machine learning models.
    models: a dict of different Scikit-learn ML models
    X_train: training data (no labels)
    X_test: testint data (no labels)
    y_train: training labels
    y_test: testing labels
    """
    #Set random seed
    np.random.seed(42)
    # Make a dictionary to keep model scores
    model_scores = {}
    # Loop through the models
    for name, model in models.items():
        # Fit the model into the data
        model.fit(X_train, y_train)
        #Evaluate the model and append its score to he model_scores
        model_scores[name] = model.score(X_test, y_test)
    return model_scores
```

```python
model_scores = fit_and_score(models, X_train, X_test, y_train, y_test)
model_scores
```

### Hyperparameter tuning

```python
# let's tune KNN

train_scores = []
test_scores = []

#Create a list of different values for n_neighbors
neighbors = range(1,21)

#Setup KNN instance
knn = KNeighborsClassifier()

# Loop through the different n_neighbors
for i in neighbors:
    knn.set_params(n_neighbors = i)

    #Fit the algorithm
    knn.fit(X_train,y_train)

    #Update the training scores list
    train_scores.append(knn.score(X_train, y_train))

    #Update the test scores list
    test_scores.append(knn.score(X_test,y_test))

```

```python
plt.plot(neighbors, train_scores, label="Train score")
plt.plot(neighbors, test_scores, label="Test score")

plt.xticks(np.arange(1,21,1))

plt.xlabel("Number of neighbors")
plt.ylabel("Model Score")
plt.legend()

print(f"Maximum KNN score on the test data: {max(test_scores)*100:.2f}% ")
```

### Hyperparameter tuning with RandomizedSearchCV

We're going to tune:
* LogisticRegression()
* RandomForestClassifier()  
... using RandomizedSearchCV

```python
# Create a hyperparameter grid for LogisticRegression

log_reg_grid = {"C" : np.logspace(-4, 4, 20),
               "solver" : ["liblinear"]
               }

# Create hyperparameter grid for RandomForestClassifier
rf_grid = {"n_estimators": np.arange(10,1000,50),
          "max_depth" : [None, 3, 5, 10],
           "min_samples_split" : np.arange(2, 20, 2),
           "min_samples_leaf": np.arange(1, 20, 2)
          }
```

Now we've got hyperparameter grid setup for each of our models, let's tine them using RandomizedSearchCV

```python
# Tune LogisticRegression

np.random.seed(42)

# Setup random hyperparameter search for LogisticRegression

rs_log_reg = RandomizedSearchCV(LogisticRegression(),
                               param_distributions = log_reg_grid,
                               cv = 5,
                               n_iter = 20,
                               verbose = True)

#Fit random hyperparameter search model for LogisticRegression
rs_log_reg.fit(X_train, y_train)


```

```python
rs_log_reg.best_params_
```

Now we've tuned LogisticRegression (with no much luck), let's do the same for RandomForestClassifier

```python
# Setup random seed

np.random.seed(42)

# Setup random hyperparameter search for RandomForestClassifier
rs_rf = RandomizedSearchCV(RandomForestClassifier(),
                           param_distributions = rf_grid,
                           cv = 5,
                           n_iter = 20,
                           verbose = True)

#Fit Random hyperparameter search model for RandomForestClassifier()
rs_rf.fit(X_train, y_train)
```

```python
sns.set(font_scale = 1.5)

def plot_conf_mat(y_test, y_preds):
    """
    Plots a nice looking confusion matrix using Seaborn's heatmap()
    """
    fig,ax = plt.subplots(figsize = (3,3))
    ax = sns.heatmap(confusion_matrix(y_test, y_preds),
                    annot = True,
                     cbar = False
                    ) 
    plt.xlabel("True label")
    plt.ylabel("Predicted label")

plot_conf_mat(y_test, y_preds)
```

### Calculate evaluation metrics using cross-validation

We're going to calculate precision, recall and f1_score of our model using cross-validation and to do so we'll be using `cross_val_score()`

```python
# Check best hyperparameters
gs_log_reg.best_params_
```

```python
# create a new classifier with best params
clf = LogisticRegression(C = 0.20433597178569418,
                        solver = "liblinear")
```

```python
# Cross-validated accuracy

cv_acc = cross_val_score(
    clf,
    X,
    y,
    cv = 5,
    scoring = "accuracy"
)
    
cv_acc
```

```python
cv_acc = np.mean(cv_acc)
cv_acc
```

```python
# Cross-validated precision

cv_precision = cross_val_score(
    clf,
    X,
    y,
    cv = 5,
    scoring = "precision"
)
    
cv_precision = np.mean(cv_precision)
cv_precision
```

```python
# Cross-validated recall

cv_recall = cross_val_score(
    clf,
    X,
    y,
    cv = 5,
    scoring = "recall"
)
    
cv_recall = np.mean(cv_recall)
cv_recall
```

```python
# Cross-validated f1-score

cv_f1 = cross_val_score(
    clf,
    X,
    y,
    cv = 5,
    scoring = "f1"
)
    
cv_f1 = np.mean(cv_f1)
cv_f1
```

```python
# Visualize cross-validated metrics
cv_metrics = pd.DataFrame({
    "Accuracy" : cv_acc,
    "Precision" : cv_precision,
    "Recall" : cv_recall,
    "F1" : cv_f1
} , index= [0] )

cv_metrics.T.plot.bar(title = "Cross-validated classification metrics" ,
                     legend = False);
```
